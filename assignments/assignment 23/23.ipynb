```python
import tkinter as tk
from tkinter import messagebox

def login():
    username = entry_user.get()
    password = entry_pass.get()

    if username == "admin" and password == "1234":
        messagebox.showinfo("Login Success", "Welcome! Login Successful")
    else:
        messagebox.showerror("Login Failed", "Invalid Username or Password")

root = tk.Tk()
root.title("Login System")
root.geometry("300x200")

tk.Label(root, text="Username").pack(pady=5)
entry_user = tk.Entry(root)
entry_user.pack()

tk.Label(root, text="Password").pack(pady=5)
entry_pass = tk.Entry(root, show="*")
entry_pass.pack()

tk.Button(root, text="Login", command=login).pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

rates = {
    "INR": 1,
    "USD": 83,
    "EUR": 90
}

def convert():
    try:
        amount = float(entry_amount.get())
        from_cur = from_currency.get()
        to_cur = to_currency.get()

        converted = amount * rates[to_cur] / rates[from_cur]
        result_label.config(text=f"Converted Amount: {converted:.2f}")
    except:
        messagebox.showerror("Error", "Please enter valid input")

root = tk.Tk()
root.title("Currency Converter")
root.geometry("350x300")

tk.Label(root, text="Amount").pack()
entry_amount = tk.Entry(root)
entry_amount.pack()

from_currency = tk.StringVar(value="INR")
to_currency = tk.StringVar(value="USD")

tk.Label(root, text="From Currency").pack()
tk.OptionMenu(root, from_currency, *rates.keys()).pack()

tk.Label(root, text="To Currency").pack()
tk.OptionMenu(root, to_currency, *rates.keys()).pack()

tk.Button(root, text="Convert", command=convert).pack(pady=15)
result_label = tk.Label(root, text="")
result_label.pack()

root.mainloop()
```

```python
import tkinter as tk

def calculate(op):
    num1 = float(entry1.get())
    num2 = float(entry2.get())

    if op == "+":
        result.set(num1 + num2)
    elif op == "-":
        result.set(num1 - num2)
    elif op == "*":
        result.set(num1 * num2)
    elif op == "/":
        result.set(num1 / num2)

root = tk.Tk()
root.title("Calculator")
root.geometry("300x250")

entry1 = tk.Entry(root)
entry1.pack(pady=5)

entry2 = tk.Entry(root)
entry2.pack(pady=5)

result = tk.StringVar()
tk.Entry(root, textvariable=result).pack(pady=5)

tk.Button(root, text="+", command=lambda: calculate("+")).pack(side="left", padx=5)
tk.Button(root, text="-", command=lambda: calculate("-")).pack(side="left", padx=5)
tk.Button(root, text="*", command=lambda: calculate("*")).pack(side="left", padx=5)
tk.Button(root, text="/", command=lambda: calculate("/")).pack(side="left", padx=5)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

questions = [
    ("Python is a ?", ["Programming Language", "Snake", "Car", "Fruit"], "Programming Language"),
    ("Which keyword is used for function?", ["func", "define", "def", "function"], "def"),
    ("Which symbol is used for comments?", ["//", "#", "/* */", "--"], "#")
]

index = 0
score = 0

def next_question():
    global index
    if index < len(questions):
        q, options, _ = questions[index]
        question_label.config(text=q)
        for i in range(4):
            radio_buttons[i].config(text=options[i], value=options[i])
        selected.set(None)
    else:
        messagebox.showinfo("Quiz Finished", f"Your Score: {score}/{len(questions)}")
        root.destroy()

def check_answer():
    global score, index
    if selected.get() == questions[index][2]:
        score += 1
        messagebox.showinfo("Correct", "Correct Answer!")
    else:
        messagebox.showerror("Wrong", "Wrong Answer!")
    index += 1
    next_question()

root = tk.Tk()
root.title("Quiz Game")
root.geometry("400x300")

question_label = tk.Label(root, text="", wraplength=350)
question_label.pack(pady=20)

selected = tk.StringVar()
radio_buttons = []

for _ in range(4):
    rb = tk.Radiobutton(root, text="", variable=selected, value="")
    rb.pack(anchor="w")
    radio_buttons.append(rb)

tk.Button(root, text="Submit", command=check_answer).pack(pady=20)

next_question()
root.mainloop()
```
