```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    op = input("Enter operator (+, -, *, /): ")

    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    elif op == '*':
        result = num1 * num2
    elif op == '/':
        if num2 == 0:
            raise ZeroDivisionError
        result = num1 / num2
    else:
        raise ValueError("Invalid Operator")

    print("Result:", result)

except ValueError as e:
    print("Error: Invalid number or operator entered")

except ZeroDivisionError:
    print("Error: Division by zero is not allowed")

except Exception as e:
    print("Unexpected error:", e)
```

```python
class Television:
    def __init__(self):
        self.model_no = 0
        self.screen_size = 0
        self.price = 0

    def accept(self):
        try:
            self.model_no = int(input("Enter model number: "))
            if self.model_no > 9999:
                raise ValueError("Model number must be 4 digits")

            self.screen_size = int(input("Enter screen size (in inches): "))
            if self.screen_size < 12 or self.screen_size > 70:
                raise ValueError("Invalid screen size")

            self.price = int(input("Enter price: "))
            if self.price < 0 or self.price > 5000:
                raise ValueError("Invalid price")

        except Exception as e:
            print("Exception occurred:", e)
            self.model_no = 0
            self.screen_size = 0
            self.price = 0

    def display(self):
        print("\nTelevision Details")
        print("Model Number:", self.model_no)
        print("Screen Size:", self.screen_size)
        print("Price:", self.price)


# main()
tv = Television()
tv.accept()
tv.display()
```
