```python
class book:
    count=0
    def __init__(self,b_id=0,name="",price=0,author=""):
        self.b_id=b_id
        self.name=name
        self.price=price
        self.author=author
        book.count+=1

    def showbook(self):
        print("id:",self.b_id)
        print("name:",self.name)
        print("price:",self.price)
        print("author:",self.author)
        

    def __del__(self):
        print("Object destroyed!")

b1=book(1,"Python",500,"Me")
b2=book(2,"Java",350,"You")

b1.showbook()
b2.showbook()
print("count of books:",book.count)
```

```python
class product:

    discount=10
    
    def __init__(self,p_id=0,name="",price=0,quantity=0):
        self.p_id=p_id
        self.name=name
        self.price=price
        self.quantity=quantity

    def applyDiscount(self):
        discount_amount=(self.price*product.discount)/100
        self.price-=discount_amount
        
    def showBook(self):
        print("id:",self.p_id)
        print("name:",self.name)
        print("price:",self.price)
        print("quantity:",self.quantity)

    def __del__(self):
        print("product object destroyed!")

p1=product(1,"laptop",60000,5)

p1.applyDiscount()
p1.showBook()
```

```python
class Shirt:
    
    increment = 10

    
    def __init__(self, sid=0, sname="", stype="", price=0.0, size=""):
        self.sid = sid
        self.sname = sname
        self.stype = stype
        self.base_price = price
        self.size = size
        self.price = self.calculatePrice()

    
    def calculatePrice(self):
        if self.size.lower() == "small":
            return self.base_price
        elif self.size.lower() == "medium":
            return self.base_price * 1.1
        elif self.size.lower() == "large":
            return self.base_price * 1.2
        elif self.size.lower() == "xlarge":
            return self.base_price * 1.3
        else:
            return self.base_price

    
    def showShirt(self):
        print("Shirt ID:", self.sid)
        print("Name:", self.sname)
        print("Type:", self.stype)
        print("Size:", self.size)
        print("Price:", self.price)

 
    def __del__(self):
        print("Shirt object destroyed")



s1 = Shirt(301, "Peter England", "Formal", 1000, "Large")
s1.showShirt()
```
