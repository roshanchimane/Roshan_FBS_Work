```python
class Student:
    def __init__(self, student_id=0, name="", age=0, percentage=0.0):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.percentage = percentage

    def accept(self):
        self.student_id = int(input("Enter Student ID: "))
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))
        self.percentage = float(input("Enter Percentage: "))

    def display(self):
        print(self)

    def calculate_rank(self):
        if self.percentage >= 85:
            return "Distinction"
        elif self.percentage >= 60:
            return "First Class"
        elif self.percentage >= 50:
            return "Second Class"
        else:
            return "Fail"

    def __str__(self):
        return (f"ID: {self.student_id}, Name: {self.name}, "
                f"Age: {self.age}, Percentage: {self.percentage}, "
                f"Rank: {self.calculate_rank()}")
```

```python
class EnggStudent(Student):
    def __init__(self, student_id=0, name="", age=0, percentage=0.0,
                 branch="", internal_marks=0):
        super().__init__(student_id, name, age, percentage)
        self.branch = branch
        self.internal_marks = internal_marks

    def accept(self):
        super().accept()
        self.branch = input("Enter Branch: ")
        self.internal_marks = int(input("Enter Internal Marks: "))

    def calculate_rank(self):
        total = self.percentage + (self.internal_marks / 2)
        if total >= 90:
            return "Excellent"
        elif total >= 70:
            return "Good"
        else:
            return "Average"

    def display(self):
        print(self)

    def __str__(self):
        return (super().__str__() +
                f", Branch: {self.branch}, Internal Marks: {self.internal_marks}")
```

```python
class College:
    def __init__(self, capacity):
        self.capacity = capacity
        self.students = []

    def add_student(self, student):
        if len(self.students) < self.capacity:
            self.students.append(student)
        else:
            print("College capacity full!")

    def get_student(self, student_id):
        for s in self.students:
            if s.student_id == student_id:
                return s
        return None

    def remove_student(self, student_id):
        self.students = [s for s in self.students if s.student_id != student_id]

    def __str__(self):
        result = "College Student List:\n"
        for s in self.students:
            result += str(s) + "\n"
        return result
```
