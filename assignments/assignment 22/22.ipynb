```python
import pickle
import os

# Employee class
class Emp:
    def __init__(self, eid, ename, basic):
        self.eid = eid
        self.ename = ename
        self.basic = basic

    def __str__(self):
        return f"ID: {self.eid}, Name: {self.ename}, Basic Salary: {self.basic}"


FILENAME = "emp.dat"


# Add record
def add_record():
    with open(FILENAME, "ab") as f:
        eid = int(input("Enter Employee ID: "))
        ename = input("Enter Employee Name: ")
        basic = float(input("Enter Basic Salary: "))
        emp = Emp(eid, ename, basic)
        pickle.dump(emp, f)
        print("‚úÖ Record added successfully")


# Display all records
def display_all():
    if not os.path.exists(FILENAME):
        print("‚ùå No records found")
        return

    with open(FILENAME, "rb") as f:
        print("\nüìã Employee Records:")
        try:
            while True:
                emp = pickle.load(f)
                print(emp)
        except EOFError:
            pass


# Search record by ID
def search_record():
    eid = int(input("Enter Employee ID to search: "))
    found = False

    with open(FILENAME, "rb") as f:
        try:
            while True:
                emp = pickle.load(f)
                if emp.eid == eid:
                    print("‚úÖ Record found:")
                    print(emp)
                    found = True
                    break
        except EOFError:
            pass

    if not found:
        print("‚ùå Record not found")


# Delete record
def delete_record():
    eid = int(input("Enter Employee ID to delete: "))
    temp = []
    found = False

    with open(FILENAME, "rb") as f:
        try:
            while True:
                emp = pickle.load(f)
                if emp.eid != eid:
                    temp.append(emp)
                else:
                    found = True
        except EOFError:
            pass

    with open(FILENAME, "wb") as f:
        for emp in temp:
            pickle.dump(emp, f)

    if found:
        print("‚úÖ Record deleted successfully")
    else:
        print("‚ùå Record not found")


# Edit record
def edit_record():
    eid = int(input("Enter Employee ID to edit: "))
    temp = []
    found = False

    with open(FILENAME, "rb") as f:
        try:
            while True:
                emp = pickle.load(f)
                if emp.eid == eid:
                    print("Enter new details:")
                    emp.ename = input("New Name: ")
                    emp.basic = float(input("New Basic Salary: "))
                    found = True
                temp.append(emp)
        except EOFError:
            pass

    with open(FILENAME, "wb") as f:
        for emp in temp:
            pickle.dump(emp, f)

    if found:
        print("‚úÖ Record updated successfully")
    else:
        print("‚ùå Record not found")


# Menu-driven program
while True:
    print("""
========= MENU =========
1. Add Record
2. Search Record
3. Delete Record
4. Edit Record
5. Display All Records
6. Exit
========================
""")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        add_record()
    elif choice == 2:
        search_record()
    elif choice == 3:
        delete_record()
    elif choice == 4:
        edit_record()
    elif choice == 5:
        display_all()
    elif choice == 6:
        print("üëã Exiting program")
        break
    else:
        print("‚ùå Invalid choice")
```
