```python
class book:
    def __init__(self,b_id=0,b_name="",price=0.0,author=""):
        self.b_id=b_id
        self.b_name=b_name,
        self.price=price,
        self.author=author
        
    def showBook(self):
        print("Book id:",self.b_id)
        print("Book Name:",self.b_name)
        print("Book price:",self.price)
        print("Book Author:",self.author)

    def __del__(self):
        print("book object deleted")

b1=book(1,"Python",100,"Guido")
b1.showBook()
```

```python
class product:
    def __init__(self,p_id=0,p_name="",p_price=0,p_quantity=0):
        self.p_id=p_id
        self.p_name=p_name
        self.p_price=p_price
        self.p_quantity=p_quantity

    def showProduct(self):
        print("Product id:",self.p_id)
        print("product name:",self.p_name)
        print("product price:",self.p_price)
        print("product quantity:",self.p_quantity)

    def __del__(self):
        print("Object deleted")
p1=product(1,"Laptop",50000,5)
p1.showProduct()
```

```python
class shirt:
    def __init__(self,s_id=0,s_name="",s_type="",s_price=0,s_size=""):
        self.s_id=s_id
        self.s_name=s_name
        self.s_type=s_type
        self.s_price=s_price
        self.s_size=s_size

    def showShirt(self):
        print("Shirt id:",self.s_id)
        print("Shirt name:",self.s_name)
        print("Shirt type:",self.s_type)
        print("Shirt price:",self.s_price)
        print("Shirt_size:",self.s_size)

    def __del__(self):
        print("shirt object deleted!")

s1=shirt(1,"USPA","Formal",2000,"Large")
s2=shirt(2,"Peter England ",1800,"Medium")
s1.showShirt()
print()
s2.showShirt()
```
