```python
lst=[2,1,3,4,2,5,1]
target=6
pairs=set()
for i in range(len(lst)):
    for j in range(i+1,len(lst)):
        if lst[i]+lst[j]==target:
            pairs.add((lst[i],lst[j]))

print(pairs)
```

```python
lst=[1,2,3,4,5,6,7,8]
key=12
pair=set()
for i in range(len(lst)):
    for j in range(i+1,len(lst)):
        for k in range(j+1,len(lst)):
            if lst[i]+lst[j]+lst[k]==key:
                pair.add(tuple(sorted((lst[i],lst[j],lst[k]))))
print(pair)
```
