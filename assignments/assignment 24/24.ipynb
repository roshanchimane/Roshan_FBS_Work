```python
import threading

total_sum = 0
lock = threading.Lock()

def sum_of_squares(start, end):
    global total_sum
    local_sum = 0
    for i in range(start, end + 1):
        local_sum += i * i

    with lock:
        total_sum += local_sum

threads = []
ranges = [(1, 25), (26, 50), (51, 75), (76, 100)]

for r in ranges:
    t = threading.Thread(target=sum_of_squares, args=r)
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Total Sum of Squares:", total_sum)
```

```python
import threading

lock = threading.Condition()
number = 1

def print_odd():
    global number
    while number <= 10:
        with lock:
            if number % 2 == 0:
                lock.wait()
            print("Odd:", number)
            number += 1
            lock.notify()

def print_even():
    global number
    while number <= 10:
        with lock:
            if number % 2 != 0:
                lock.wait()
            print("Even:", number)
            number += 1
            lock.notify()

t1 = threading.Thread(target=print_odd)
t2 = threading.Thread(target=print_even)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading
import time

def print_lowercase():
    for c in range(ord('a'), ord('z') + 1):
        print(chr(c), end=" ")
        time.sleep(0.1)

def print_uppercase():
    for c in range(ord('A'), ord('Z') + 1):
        print(chr(c), end=" ")
        time.sleep(0.1)

t1 = threading.Thread(target=print_lowercase)
t2 = threading.Thread(target=print_uppercase)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading
import time
import random

buffer = []
buffer_size = 5
lock = threading.Condition()

def producer(pid):
    while True:
        item = random.randint(1, 100)
        with lock:
            while len(buffer) == buffer_size:
                lock.wait()
            buffer.append(item)
            print(f"Producer {pid} produced {item}")
            lock.notify()
        time.sleep(1)

def consumer(cid):
    while True:
        with lock:
            while len(buffer) == 0:
                lock.wait()
            item = buffer.pop(0)
            print(f"Consumer {cid} consumed {item}")
            lock.notify()
        time.sleep(1)

producers = [
    threading.Thread(target=producer, args=(1,)),
    threading.Thread(target=producer, args=(2,))
]

consumers = [
    threading.Thread(target=consumer, args=(1,)),
    threading.Thread(target=consumer, args=(2,))
]

for p in producers:
    p.start()

for c in consumers:
    c.start()
```
