```python
class SYMARKS:
    def __init__(self, computer, maths, electronics):
        self.computer = computer
        self.maths = maths
        self.electronics = electronics


class TYMarks:
    def __init__(self, theory, practical):
        self.theory = theory
        self.practical = practical


class Student:
    def __init__(self, roll, name, sy, ty):
        self.roll = roll
        self.name = name
        self.sy = sy
        self.ty = ty

    def display(self):
        total = self.sy.computer + self.ty.theory

        if total >= 70:
            grade = "A"
        elif total >= 60:
            grade = "B"
        elif total >= 50:
            grade = "C"
        elif total >= 40:
            grade = "Pass Class"
        else:
            grade = "Fail"

        print("Roll No:", self.roll)
        print("Name:", self.name)
        print("Total:", total)
        print("Grade:", grade)


sy = SYMARKS(35, 60, 55)
ty = TYMarks(40, 45)

s = Student(1, "Roshan", sy, ty)
s.display()
```
