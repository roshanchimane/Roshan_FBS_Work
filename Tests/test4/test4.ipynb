```python
def print_factors(n):
    print(f"factors of {n} are:")
    for i in range(1,n+1):
        if n%i==0:
            print(i,end=" ")
num=int(input("enter num:"))
print_factors(num)
```

```python
# recursion

def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*(factorial(n-1))

n=4
factorial(n)
```

```python
def print_pattern():
    n = 9

    # Top horizontal line
    for i in range(n):
        print("*", end=" ")
    print()

    # Diagonal
    for i in range(1, n - 1):
        for j in range(n-i-1):
            print(" ", end=" ")
        print("*")

    # Bottom horizontal line
    for i in range(n):
        print("*", end=" ")

print_pattern()
```
